```python
import matplotlib.pyplot as plt
import json

fig, axs = plt.subplots(1, 2)
fig.set_size_inches(18, 6)
axs[0].grid()
axs[1].grid()
for i, fname in enumerate([
    "./result/distribution_learning/result_best.json", 
    "./result/hill_climb/result_003.json", 
    "./result/hill_climb/result_007.json", 
    "./result/hill_climb/result_011.json", 
    "./result/hill_climb/result_015.json", 
    "./result/hill_climb/result_019.json"
    ]):
    with open(f"{fname}", "r") as f:
        result = json.load(f)
    
    axs[0].scatter(result["logp"], result["sa"], s=5, alpha=0.3)
    axs[1].scatter(result["logp"], result["qed"], s=5, alpha=0.3)
    #axs[2].scatter(result["penalized_logp"], result["rd_filter"], s=3)

plt.savefig("./result/hill_climb/tradoff.png")
plt.show()
```
